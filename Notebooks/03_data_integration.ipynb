```python
# ===========================
# 1. load cleaned datasets
# ===========================

import pandas as pd

gdp = pd.read_csv('../Data/Processed/gdp_clean.csv')
sales = pd.read_csv('../Data/Processed/sales_clean.csv')

# ensure correct datatypes
gdp['year'] = gdp['year'].astype(int)
sales['year'] = sales['year'].astype(int)
```

```python
# ============================
# aggregate sales by year country and indicator
# ============================

sales_clean = sales.groupby(['country', 'year', 'indicator'], as_index=False)['sales_value'].mean()
final_df = sales_clean.merge(gdp, on=["country", "year"], how="inner")
```

```python
# ============================
# rename columns
# ============================

final_df = sales_clean.merge(gdp, on=["country", "year"], how="inner")

# rename GDP unit column
final_df = final_df.rename(columns={
    "unit": "gdp_unit"
})

# data cleaning, aggregates removal
aggregates = ["EA19", "EA20", "EA21", "EU27_2020"]
final_df = final_df[~final_df["country"].isin(aggregates)]
```

```python
# ============================
# Check overlaps
# ============================

print("GDP dataset years:", gdp['year'].unique())
print("Sales dataset years:", sales_clean['year'].unique())

print("GDP dataset countries:", gdp['country'].unique())
print("Sales dataset countries:", sales_clean['country'].unique())
```

```python
# ================================
# 3. merge datasets
# ================================

final_df = sales_clean.merge(gdp, on=['country', 'year'], how='inner')
```

```python
# =============================
# Check the merged dataset
# ================================

# check the range of years
print("Year range:", final_df["year"].min(), "-", final_df["year"].max())

print("Final dataset shape:", final_df.shape)
print("Final dataset columns:", final_df.columns)
print("Final dataset sample:")
print(final_df.head())
print("Final dataset years:", final_df['year'].unique())
print("Final dataset countries:", final_df['country'].unique())
print("Final dataset indicators:", final_df['indicator'].unique())
print(final_df.describe())
print(final_df.info())
```

```python
# ==============================
# 6. save and export the data
# ==============================

final_df.to_csv('../Data/Processed/final_dataset.csv', index=False)
```
